```python
#a.Data Loading
import pandas as pd
data = pd.read_csv("student-mat.csv")
print(data.head())
```

```python
#b.Data Exploration
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)
print("Data Types:\n", data.dtypes)
print("Dataset Shape:", data.shape)
```

```python
#c.Data Cleaning
numeric_columns = data.select_dtypes(include='number').columns
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].median())

data.drop_duplicates(inplace=True)
print("Cleaned Dataset Shape:", data.shape)
```

```python
#d.Data Analysis
import pandas as pd
data = pd.read_csv('student-mat.csv', delimiter=';')
if 'G3' in data.columns:
    average_g3 = data['G3'].mean()
    print("\nAverage score in final grade (G3):", average_g3)
else:
    print("'G3' column not found. Available columns are:", data.columns)
```

```python
above_15_count = data[data['G3'] > 15].shape[0]
print("Students scoring above 15:", above_15_count)
```

```python
correlation = data['studytime'].corr(data['G3'])
print("Correlation between study time and G3:", correlation)
```

```python
gender_avg = data.groupby('sex')['G3'].mean()
print("Average Final Grade by Gender:\n", gender_avg)
```

```python
#e.Data Visualization
import matplotlib.pyplot as plt

plt.hist(data['G3'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of Final Grades (G3)')
plt.xlabel('Final Grades (G3)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(data['studytime'], data['G3'], alpha=0.7, color='purple')
plt.title('Study Time vs Final Grade (G3)')
plt.xlabel('Study Time')
plt.ylabel('Final Grade (G3)')
plt.show()
```

```python
gender_avg.plot(kind='bar', color=['blue', 'pink'])
plt.title('Average Final Grade by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Final Grade (G3)')
plt.xticks(rotation=0)
plt.show()
```
